```python
import os
import pandas as pd

# Directory path containing the CSV files
directory = '/Users/garfieldgreglim/Downloads/csvs'

# List to store the loaded DataFrames
dfs = []

# Iterate over the files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):  # Ensure only CSV files are processed
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath)
        dfs.append(df)

# Now you have a list of loaded DataFrames (dfs)
# You can perform further operations on the DataFrames as needed
```

```python
import pandas as pd

merged_df = pd.concat(dfs)
df = merged_df
df = df.rename(columns=lambda x: x.lower())
df
```

```python
import uuid
import pandas as pd

# Assume df is your DataFrame

# Generate a UUID for each row and create a new 'uuid' column
df['uuid'] = [str(uuid.uuid4()) for _ in range(len(df))]

# Move the 'uuid' column to the first position
uuid_col = df.pop('uuid')
df.insert(0, 'uuid', uuid_col)
df
```

```python
df = pd.melt(df, id_vars=['uuid'], value_vars=['description', 'aims', 'scope', 'editorial process', 'peer review process', 'criteria for acceptance and rejection', 'comments', 'transparency', 'privacy statement', 'managing editor', 'associate editors', 'editorial support staff', 'recruiting referees', 'authorship and contributorship', 'plagiarism', 'duplicate and redundant policy', 'policy on conflicts of interest and funding', 'open access and copyright policy', 'appeals and complaints'])
df = df.drop(columns=['variable'])
df.columns = ['content_id', 'content']
df = df.dropna()
df
```

```python
import uuid
import pandas as pd

# Assume df is your DataFrame

# Generate a UUID for each row and create a new 'uuid' column
df['uuid'] = [str(uuid.uuid4()) for _ in range(len(df))]

# Move the 'uuid' column to the first position
uuid_col = df.pop('uuid')
df.insert(0, 'uuid', uuid_col)
df
```

```python
data_list = []

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Create a dictionary for each row
    row_dict = {"uuid": row["uuid"], "object": {"content_id": row["content_id"], "content": row["content"]}, "content": str(row['content'])} 
    
    # Add each dictionary to the list
    data_list.append(row_dict)
```

```python
data_list = []

# Iterate over DataFrame rows
for index, row in df.iterrows():
    # Create a tuple for each row and append it to the list
    data_list.append((row["uuid"], {"content_id": row["content_id"], "text": row["content"]}, str(row['content'])))
```

```python
data_list[0][2]
dict_list = data_list
```

```python
new_dict_list = []

for dic in dict_list:
    new_string_repr = dic[2].replace('{', '').replace('}', '').replace("\n", "").replace("\\", "")
    new_dict_entry = (dic[0], dic[1], new_string_repr)
    new_dict_list.append(new_dict_entry)
dict_list = new_dict_list
```
