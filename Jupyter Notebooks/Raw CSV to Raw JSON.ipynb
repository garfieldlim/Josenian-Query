```python
csv_folder = "raw_csvs/"
csv_file = "post.csv"
json_folder = "raw_jsons/"

file_path = csv_folder + csv_file

# Read the CSV file, specifying that the first row contains column names
df = pd.read_csv(file_path, header=0, encoding='latin1')

df
```

Dropping

```python
# Drop the last column
# # 4. Rename the 'post_id' column to 'uuid'
df.rename(columns={'post_id': 'uuid', 'ocr_image ': 'ocr',}, inplace=True)

# 5. Generate new uuids
df['uuid'] = [str(uuid.uuid4()) for _ in range(len(df))]

df

df = df.iloc[:, :-1]
df['link'] = df['video'].astype(str) + df['image'].astype(str)

# 1. Remove 'image_lowquality' column


# 3. Remove duplicates based on 'post_id'
df.drop_duplicates(subset='uuid', keep='first', inplace=True)

# df['link'] = df['link'].replace('nannan', np.nan)

# Concatenate 'text' and 'ocr' columns into a new column 'combined'
df['combined'] = df['text'] + df['ocr'].fillna('')

df.drop(columns=['image_lowquality', 'video', 'image', 'link', 'ocr', 'text'], inplace=True)
df.rename(columns={'combined': 'text',}, inplace=True)
display(df)
```

```python
# This list will contain your dictionaries
dict_list = []

# Iterate over the DataFrame rows
for idx, row in df.iterrows():
    # Create a sub dictionary with each column name as a key and the corresponding value from the row
    sub_dict = {name: row[name] for name in df.columns}
    # Create the final tuple adding the 'text' key with the joined string values from the row
    final_tuple = (row['uuid'], sub_dict, ' '.join(f'{name}: {row[name]},' for name in df.columns))
    # Append the tuple to the list
    dict_list.append(final_tuple)
```

```python
from datetime import datetime

def change_time_format(time_string):
    dt = datetime.strptime(time_string, "%Y-%m-%dT%H:%M:%S")
    formatted_time = dt.strftime("%Y-%m-%d %B %d, %Y %H:%M:%S")
    return formatted_time
```

```python
import requests
import pandas as pd
from tqdm import tqdm
import openai

# OpenAI setup
openai.api_key = 'your_openai_api_key'
engine = 'text-davinci-004'

# Wrap your iteration with tqdm for progress bar
for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    # Check if the 'image' column is not NaN
    if pd.notnull(row['image']):
        # Pass the image URL to the OCR API
        ocr_response = requests.post('https://api.ocr.space/parse/image', 
            data={'apikey': 'helloworld', 'url': row['image']})

        if ocr_response.status_code == 200:
            # Error handling for OCR API response
            ocr_json = ocr_response.json()
            if 'ParsedResults' in ocr_json:
                # Get the OCR text
                ocr_text = ocr_json['ParsedResults'][0]['ParsedText']

                # Use OpenAI's API to correct the grammar
                prompt = f"Please correct the following text:\n\n{ocr_text}"
                response = openai.Completion.create(engine=engine, prompt=prompt, max_tokens=500)

                # Get the corrected text and store it in the 'ocr' column
                corrected_text = response.choices[0].text.strip()
                df.at[index, 'ocr'] = corrected_text
                print(corrected_text)
            else:
                print(f"OCR error for index {index}, image {row['image']}: {ocr_json.get('ErrorMessage', 'Unknown error')}")
```
