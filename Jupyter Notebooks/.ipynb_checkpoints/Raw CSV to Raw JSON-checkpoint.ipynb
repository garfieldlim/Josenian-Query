```python
path_folder = "raw_csvs/"
csv_file = "posts.csv"

file_path = path_folder + csv_file

df = pd.read_csv(file_path)

display(df)
```

```python
import pandas as pd
import numpy as np

# 1. Remove 'image_lowquality' column
df = df.drop(columns=['image_lowquality'])

# 2. Replace "How can" and "How may" in 'ocr' column with NaN
df.loc[df['ocr'].str.contains("How can|How may|I'm sorry|You're welcome|Unfortunately|Sorry|to help|how can", na=False), 'ocr'] = np.nan

# # 3. Sorting into different dataframes


# # Remove rows where 'ocr' is not NaN and 'image' and 'video' are NaN
df = df.loc[~((df['ocr'].notna()) & (df['image'].isna()) & (df['video'].isna()))]

df_with_image_with_ocr = df.dropna(subset=['image', 'ocr'])
df_with_image_only = df.dropna(subset=['image']).loc[df['ocr'].isna()]
df_with_video_only = df.dropna(subset=['video']).loc[df['ocr'].isna()]
df_with_video_with_ocr = df.dropna(subset=['video', 'ocr'])
```

```python
import requests
import pandas as pd
from tqdm import tqdm
import openai

# OpenAI setup
openai.api_key = 'your_openai_api_key'
engine = 'text-davinci-004'

# Wrap your iteration with tqdm for progress bar
for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    # Check if the 'image' column is not NaN
    if pd.notnull(row['image']):
        # Pass the image URL to the OCR API
        ocr_response = requests.post('https://api.ocr.space/parse/image', 
            data={'apikey': 'helloworld', 'url': row['image']})

        if ocr_response.status_code == 200:
            # Error handling for OCR API response
            ocr_json = ocr_response.json()
            if 'ParsedResults' in ocr_json:
                # Get the OCR text
                ocr_text = ocr_json['ParsedResults'][0]['ParsedText']

                # Use OpenAI's API to correct the grammar
                prompt = f"Please correct the following text:\n\n{ocr_text}"
                response = openai.Completion.create(engine=engine, prompt=prompt, max_tokens=500)

                # Get the corrected text and store it in the 'ocr' column
                corrected_text = response.choices[0].text.strip()
                df.at[index, 'ocr'] = corrected_text
                print(corrected_text)
            else:
                print(f"OCR error for index {index}, image {row['image']}: {ocr_json.get('ErrorMessage', 'Unknown error')}")
```
