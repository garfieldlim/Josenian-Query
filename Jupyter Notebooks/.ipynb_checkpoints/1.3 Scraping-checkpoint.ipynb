```python
import requests
from bs4 import BeautifulSoup
import csv
import os
import urllib.parse

# Define the base directory where you want to save the folders
base_dir = "./1.3 Scraped Data/"

# Define the extensions of media files you want to check for
media_extensions = ['.mp4', '.jpg', '.png', '.gif', '.webm', '.ogg', '.mp3', '.wav', '.flac']

# Open the CSV file
with open("1.2 Files/dfs/df_3.csv", 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    # Loop over each row in the CSV file
    for row in reader:
        # Loop over each URL in the row
        for url in row:
            if not url:  # skip empty strings
                continue
            # Parse the URL and check if it's a media file
            print(url)
            parsed_url = urllib.parse.urlparse(url)
            if any(parsed_url.path.endswith(ext) for ext in media_extensions):
                # If it's a media file, write the URL to a text file in the corresponding folder
                dir_name = urllib.parse.quote_plus(url)
                dir_path = os.path.join(base_dir, dir_name)
                os.makedirs(dir_path, exist_ok=True)
                with open(os.path.join(dir_path, 'media.txt'), 'a') as f:
                    f.write(url + '\n')
                continue  # Skip the rest of the loop for media files

            # Request the webpage
            response = requests.get(url)
            if response.status_code == 200:
                soup = BeautifulSoup(response.text, 'html.parser')
                # Kill all script and style elements
                for script in soup(["script", "style"]):
                    script.decompose()    # rip it out
                
                # Get text
                text = soup.get_text()
                # Break into lines and remove leading and trailing space on each
                lines = (line.strip() for line in text.splitlines())
                # Break multi-headlines into a line each
                chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
                # Drop blank lines
                text = '\n'.join(chunk for chunk in chunks if chunk)

                # Make a safe directory name by removing/encoding special characters from the URL
                dir_name = urllib.parse.quote_plus(url)
                dir_path = os.path.join(base_dir, dir_name)
                os.makedirs(dir_path, exist_ok=True)
                
                # Write the content to a file in the directory
                with open(os.path.join(dir_path, 'content.txt'), 'w') as f:
                    f.write(url + '\n' + text)

                # Extract and write image URLs
                with open(os.path.join(dir_path, 'image.txt'), 'w') as f:
                    for img in soup.find_all('img'):
                        f.write(img.get('src') + '\n')

                # Extract and write video URLs
                with open(os.path.join(dir_path, 'video.txt'), 'w') as f:
                    for video in soup.find_all('video'):
                        for source in video.find_all('source'):
                            f.write(source.get('src') + '\n')

                # Extract and write audio URLs
                with open(os.path.join(dir_path, 'audio.txt'), 'w') as f:
                    for audio in soup.find_all('audio'):
                        for source in audio.find_all('source'):
                            f.write(source.get('src') + '\n')

            else:
                print(f"Failed to retrieve page at {url}, status code: {response.status_code}")
```
