```python
import os
import pandas as pd

# specify the directory you want to scan for CSV files.
csv_folder = 'RMRJ'  # update this to your specific path if required

# use os.listdir to get all files in the directory and sort them
files = sorted(os.listdir(csv_folder))

# process the sorted files
for filename in files:
    # only process .csv files
    if filename.endswith('.csv'):
        # construct full file path
        filepath = os.path.join(csv_folder, filename)
        
        # load the CSV file as a pandas DataFrame
        df = pd.read_csv(filepath)
        
        # print the filename and the DataFrame
        print(f'File: {filename}')
        display(df)
        print()
```
